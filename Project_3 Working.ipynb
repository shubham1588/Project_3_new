#### ●	Univariate analysis is the simplest form of analyzing data.
#### ●	“Uni” means “one”, so in other words, data has only one variable.

#### ●	It doesn’t deal with causes or relationships and its major purpose is to describe. It takes data, summarizes that data, and finds patterns in the data.
#### ●	Univariate Analysis can be done either on numerical or categorical features.

## Distribution plot "distplot"
### We have taken major numberic variable to find the data distribution in these variables 

```python
Number_Column=["YearBuilt","TotalBsmtSF","GrLivArea","SalePrice"]

for i in range(0,len(Number_Column),2):
    plt.figure(figsize=(8,4))
    plt.subplot(121)
    sns.distplot(df[Number_Column[i]],kde=False)
    plt.subplot(122)
    sns.distplot(df[Number_Column[i+1]],kde=False)
    plt.tight_layout()
    plt.show()
```
